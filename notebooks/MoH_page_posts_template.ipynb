```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.set()
%config InlineBackend.figure_format = 'retina'
```

```python
posts = pd.read_csv('../data/2020-11-23-MoH-posts.csv')
```

```python
posts.head()
```

## Interaction and sentiment stats for posts

```python
posts.sort_values(by='Created', inplace=True)
posts.head()
```

```python
posts['Created'] = posts['Created'].astype('datetime64')
posts['Created'] = posts['Created'].dt.strftime('%b %d %Y')
```

```python
posts.groupby(by='Type').count()
```

### Likes

```python
posts[['Message', 'Created', 'Likes']].sort_values(by='Likes', ascending=False).head()
```

```python
ax = posts.plot(x='Created', y='Likes', color='Blue', alpha=0.6,  figsize=(10,5))
ax.set(title="Likes on MOH Facebook posts")
ax.set(ylabel="Likes")
plt.show()
```

### Angry reactions

```python
posts[['Message', 'Created', 'Angry']].sort_values(by='Angry', ascending=False).head(5)
```

```python
ax = posts.plot(x='Created', y='Angry', color='Red', alpha=0.6, figsize=(10,5))
ax.set(title="Angry reactions on MOH Facebook posts")
ax.set(ylabel="Angry reactions")
```
