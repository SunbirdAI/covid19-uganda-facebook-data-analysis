## Annotation

### Prepare posts for annotation

```python
posts_for_annotation = posts[['Facebook Id','Message']]
posts_for_annotation.rename(columns={"Message": "text"}, inplace=True)
posts_for_annotation['label'] = ''
```

```python
posts_for_annotation.to_csv('MoH-posts-for-annotation-4.csv')
```

### Analyze annotated posts

```python
annotated_posts = pd.read_csv('../data/annotated_MoH_posts_4.csv')
annotated_posts.head(10)
```

```python
annotated_posts['category'] = annotated_posts['label'].astype('category')
```

```python
annotated_posts['category'].cat.categories = ['New COVID cases or deaths', 'Other COVID information', 'Non-COVID information']
annotated_posts.head(10)
```

```python
sns.countplot(y='category', data=annotated_posts)
```

```python
final_df = annotated_posts[['text', 'category']]
final_df.head(5)
```

```python
final_df.to_csv('MoH-posts-text-4.csv')
```
