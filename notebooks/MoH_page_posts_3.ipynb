```python
import pandas as pd
import seaborn as sns
```

```python
posts = pd.read_csv('../data/2020-10-07-MoH-posts.csv')
```

```python
posts.head()
```

## Interaction and sentiment stats for posts

```python
posts.sort_values(by='Created', ascending=False, inplace=True)
posts.head()
```

### Likes

```python
posts[['Message', 'Created', 'Likes']].sort_values(by='Likes', ascending=False).head()
```

```python
posts.plot.area(x='Created', y='Likes', color='Green', alpha=0.4,  figsize=(20,10))
```

### Angry reactions

```python
posts[['Message', 'Created', 'Angry']].sort_values(by='Angry', ascending=False).head(11)
```

```python
posts.plot.area(x='Created', y='Angry', color='Red', alpha=0.4,  figsize=(20,10))
```

## Annotation

### Prepare posts for annotation

```python
posts_for_annotation = posts[['Facebook Id', 'Message']]
posts_for_annotation.rename(columns={"Message": "text"}, inplace=True)
posts_for_annotation['label'] = ''
```

```python
posts_for_annotation.to_csv('MoH-posts-for-annotation-3.csv')
```

### Analyze annotated posts

```python
annotated_posts = pd.read_csv('../data/annotated_MoH_posts_3.csv')
annotated_posts.head(10)
```

```python
annotated_posts['category'] = annotated_posts['label'].astype('category')
annotated_posts['category']
```

```python
annotated_posts['category'].cat.categories = ['New cases or deaths', 'Other COVID information', 'Non-COVID information']
annotated_posts.head(10)
```

```python
sns.countplot(y='category', data=annotated_posts)
```
