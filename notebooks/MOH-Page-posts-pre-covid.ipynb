## Preparation of MOH Page posts from pre-COVID times 

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
posts = pd.read_csv('../data/MoH-Page-posts-2014-11-29--2019-11-30.csv')
```

```python
posts.head(5)
```

```python
posts = posts[['Message', 'Description']]
# posts.dropna(subset=['Message', 'Description'], inplace=True)
# posts.count()
```

```python
posts['Message'].fillna(posts['Description'], inplace=True)
posts.rename(columns={"Message": "text"}, errors="raise", inplace=True)
posts.drop(columns=['Description'], axis=1, inplace=True)
posts['category'] = 'Non-COVID'
posts.dropna(inplace=True)
```
